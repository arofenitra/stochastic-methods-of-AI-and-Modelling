## Probability Space
It is a probability triple $(\Sigma,F,P)$, where :
- $\Sigma$ is a sample space or set of possible outcome or universe of possibilities.
- $F$ is the event space, known also as sigma-algebra. This is a collection of some subsets of $\Sigma$ which includes $∅$ and $\Sigma$, and countable union of events is an event in $F$ also.
- $P$ is a probability function which assign each event space a real numbers in $[0,1]$ such that $\forall A\in F: P(A)\in [0,1]$, $P(\Sigma)=1=1-P(∅)$ and probability of countable union of mutually exclusive event is the sum of their individual probabilities.

**Example**:
Suppose we throw a 6 sided dice such that each face has an equal probability of landing face up. We have:
$$\Sigma=\{1,2,3,4,5,6\}$$
$$ F=\{∅,\{1\},...,\{6\} ,...,\{1,...,6\} \}=∪_{A\in F}A$$
$$P:F→[0,1] : P(∅)=0, P(\{i\})=\frac{1}{6} \forall i\in \{1,...,6\} , P(\{1,...,6\})=1$$


```python


#define the sample space
Sigma=np.array([1,2,3,4,5,6])

#function to roll a dice:
def roll_dice():
  return np.random.choice(Sigma)

#simulation of rolling dice n times
n_roll=10**5
results=np.array([roll_dice() for _ in range(n_roll)])

#calculate the probability based on the above results
def calculate_probability(event):
  if event==[]:
    return 0
  elif event=="sigma":
    return 1

  Sum=0
  dic_event=set()
  for x in event:
    if 1<=x and x<=6:
      dic_event.add(x)
    else:
      pass

  return len(dic_event)/6

events=[[],[1],[2],[3],[4],[5],[6],"sigma"]
probabilities=np.array([calculate_probability(event) for event in events])
for i in range(len(events)):
  print(f"Event:{events[i]}, Probability: {probabilities[i]:.4f}")
x=np.array(range(1,len(events)+1))
plt.xticks(x,events)
plt.xlabel("event")
plt.ylabel("probabilities")
plt.plot(x,probabilities,"o")
plt.title("probability of events in rolling dice")


plt.show()
```

## Distributions and characterisation
**Moments**

$P(x)$ is a probability distribution function, , the mean is $\mathbb{E}[X]=\mu$, the variance is $\sigma^2=Var(X)$, the Standard deviation is $\sigma=\sqrt{Var(X)}=\sqrt{\sigma^2}$. For each $n\in \mathbb{N}$, we have the following:
- Raw moments : $\displaystyle \langle X^n⟩=\int_{\mathbb{R}}x^nP(x)dx$.
- Central moments: $\displaystyle μ_n=\langle \left( X-\mathbb{E}[X] \right)^n⟩=\int_{\mathbb{R}}\left( x-\mathbb{E}[X] \right)^nP(x)dx $.
- Standardized moment: $\displaystyle \frac{μ_n}{\sigma^n}=⟨ \left( \frac{X-\mathbb{E}[X]}{σ} \right)^n
 ⟩=\int_{\mathbb{R}}\left( \frac{x-\mu}{σ} \right)^nP(x)dx $.


Example of moments and central moments:
- Mean or expectation Value : $\mu=\mathbb{E}[X]=\displaystyle \langle X⟩=\int_{\mathbb{R}}xP(x)dx$.
- Variance : $\sigma^2=Var(X)=\displaystyle \langle \left( X-μ \right)^n⟩=\int_{\mathbb{R}}\left( x-μ \right)^2P(x)dx$.
- Skewness : $\displaystyle \frac{\mu_3}{\sigma^3}=⟨ \left( \frac{X-μ}{σ} \right)^3
 ⟩=\int_{\mathbb{R}}\left( x-μ \right)^3P(x)dx $.
- Kurtosis :$\displaystyle \frac{\mu_4}{\sigma^4}= ⟨ \left( \frac{X-μ}{σ}\right)^4 ⟩=\int_{\mathbb{R}}\left( x-μ \right)^4P(x)dx $.

**Characteristic function**

The characteristic function of a random variable $X$ is $\displaystyle \varphi_X(t)=\mathbb{E}[e^{itX}]=\int_{\mathbb{R}}e^{itx}P(x)dx$. $φ_X(0)=\int_{\mathbb{R}}P(x)dx=1$, $\displaystyle |φ_X(t)|=\left|\int_{\mathbb{R}}e^{itx}P(x)dx \right|\leq \int_{\mathbb{R}}|e^{itx}||P(x)|dx=1$.  
$$φ_X(t)=\int_{\mathbb{R}}e^{itx}P(x)dx=\sum_{n=0}^∞\frac{(it)^n}{n!}\left(\int_{\mathbb{R}}x^nP(x)dx\right)=\sum_{n=0}^∞i^n\langle X^n ⟩\frac{t^n}{n!} $$
$$⟨X^n⟩=\frac{1}{i^n}φ_X^{(n)}(0)$$
**Moment-generating function**

The moment-generating function is $M_X(t)=\varphi_X(-it)=\mathbb{E}[e^{tX}]$

**Cumulants and cumulant-generating function**

The cumulant generating function is $\displaystyle K(t)=\log{(M_X(t))}=\log{(\mathbb{E}[e^{tX}])}=\sum_{n=1}^\infty\kappa_n\frac{t^n}{n!}$. where $\displaystyle \kappa_n= K^{(n)}(0)$ is the cumulants.  
Some author choose the cumulant generating function as $\displaystyle H(t)=\log{(\varphi_X(t))}=\sum_{n=1}^\infty\kappa_n\frac{i^n t^n}{n!}$, in that case the cumulants : $\displaystyle κ_n=\frac{1}{i^n}H^{(n)}(0)$.  

We can see that $κ_1=\mu , κ_2=\sigma^2, κ_3=\mathbb{E}[(X-μ)^3], κ_4=\mathbb{E}[(X-\mu)^4]-3κ_2^2$

**Usefull function** :
- For some $A⊆\mathbb{R}$, the [Indicator function](https://en.wikipedia.org/wiki/Indicator_function) $\chi_A:\mathbb{R} →\{0,1\};\chi_A(x)=1$ if $x\in A$ and $\chi_A(x)=0$ otherwise.





### Meaning of variance, skewness and kurtosis :
1) Variance and Standard Deviation: These measures describe the spread of the data. A wider histogram indicates higher variance.

2) Skewness:
- Positive Skewness (Skewness > 0):
The distribution has a longer or fatter tail on the right side.  
Most of the values cluster towards the left, with a few larger values pulling the mean to the right.
- Negative Skewness (Skewness < 0):
The distribution has a longer or fatter tail on the left side.  
Most of the values cluster towards the right, with a few smaller values pulling the mean to the left.  

- Zero Skewness (Skewness ≈ 0):
The distribution is approximately symmetric arround the mean.  

3) Kurtosis:
- Leptokurtic (Kurtosis > 3):
The distribution has heavier tails and a sharper peak than a normal distribution.
There is a higher probability of extreme values (outliers).
- Mesokurtic (Kurtosis = 3):
The distribution has tails and a peak similar to a normal distribution.   
Example: Standard normal distribution.
- Platykurtic (Kurtosis < 3):
The distribution has lighter tails and a flatter peak than a normal distribution.

### Bernouilli distribution
- parameters : $0\leq p \leq 1 $
- probability mass function : $P(X=1)=p \; ; \;P(X=0)=1-p$. Hence $P(x)=p χ_{\{1\}}(x)+(1-p)χ_{\{0\}}(x)$
- Cumulative density function (CDF) :  $F(x)=P(X< x)=0$ if $x\leq 0$, $P(X<x)=1-p$ if $0 \leq x \leq 1$ and $P(X<x)=1$  if  $x>1$. So $F(x)=(1-p)χ_{[0,1]}(x)+χ_{(1,∞)}(x)$.
- Expected value: $\mathbb{E}[X]=\sum_{i=0}^1iP(X=i)=0(1-p)+1(p)=p $
- Variance:
$$Var(X)=\mathbb{E}[X^2]-\mathbb{E}[X]^2=\sum_{i=0}^1i^2P(X=i)-p^2=p(1-p)$$
- Skewness: $\sigma=\sqrt{Var(X)}=\sqrt{p(1-p)}$.  Hence we have:
$$\frac{\mu_3}{\sigma^3}=\mathbb{E}[(X-\mu)^3/\sigma^3]=\int_{\mathbb{R}}\frac{(x-p)^3}{p^{3/2}(1-p)^{3/2}}\biggr(p\chi_1(x)+(1-p)\chi_0(x) \biggl)dx=p^{-3/2}(1-p)^{-3/2}\biggl[p(1-p)^3-(1-p)p^3 \biggr]=\frac{1-2p}{\sqrt{p(1-p)}}$$
- Kurtosis:
$$\frac{\mu_4}{\sigma^4}=\mathbb{E}[(X-\mu)^4/\sigma^4]=\int_{\mathbb{R}}\frac{(x-p)^4}{p^2(1-p)^2}\biggr(p\chi_1(x)+(1-p)\chi_0(x) \biggl)dx=p^{-2}(1-p)^{-2}\biggl[p(1-p)^4-(1-p)p^4 \biggr]=\frac{(p^2-p+1)(1-2p)}{p(1-p)}$$

- Characteristic function:
$$φ_X(t)=\mathbb{E}[e^{itX}]=\int_{\mathbb{R}}e^{itx}P(x)dx=p\int_{\mathbb{R}}e^{itx}\chi_{1}(x)dx+(1-p)\int_{\mathbb{R}}e^{itx}\chi_{0}(x)dx=pe^{it}+(1-p)=1+\sum_{n\geq 1}p\frac{(it)^n}{n!}$$


- Moments:
$$⟨X^n⟩=\frac{1}{i^n}\frac{∂^n}{∂t^n}(φ_X(t))\biggr|_{t=0}=\frac{∂^n}{∂(it)^n}\biggl(pe^{it}+(1-p)\biggr)\biggr|_{t=0}=p$$
- Cumulants and cumulant-generating function $\displaystyle H(t)=\log(φ_X(t))=\log(pe^{it}+(1-p))=\sum_{n\geq 0}κ_n i^n\frac{t^n}{n!}=ipt+p(1-p)\frac{(it)^2}{2!}+...$. Hence $κ_1=p, κ_2=p(1-p),...$ where $\displaystyle κ_{n+1}=p(1-p)\frac{dκ_n}{dp}$.





```python
#Bernouilli probability Density function of parameter p
def bernouilli_pdf(x,p):
  if x==0:
    return 1-p
  elif x==1:
    return p
  else:
    return 0

#Bernouilli Cumulative Distribution function of parameter p P(X<=x)
def bernouilli_cdf(x,p):
  if x <0:
    return 0
  elif x==1:
    return 1-p
  elif x>1:
    return 1

def bernouilli_mean(p):
  return p
def bernouilli_variance(p):
  return p*(1-p)
def bernouilli_skewness(p):
  return (1-2*p)/np.sqrt(p*(1-p))
def bernouilli_kurtosis(p):
  return (p**2-p+1)*(1-2*p)/(p*(1-p))

p=0.6
print("mean is : ", bernouilli_mean(p))
print("variance is : ", bernouilli_variance(p))
print("skewness is : ", bernouilli_skewness(p))
print("kurtosis is : ", bernouilli_kurtosis(p))
plt.axvline(x=0,ymin=0, ymax=1-p, color='b', linestyle='-', label='probability')
plt.axvline(x=1,ymin=0, ymax=p, color='b', linestyle='-')
plt.axvline(x=bernouilli_mean(p), color='r', linestyle='--', label='Mean')
plt.legend()
plt.xlim(xmin=-0.1,xmax=1.1)

plt.show()

#with histogram
data = np.random.binomial(1, p, size=1000)  # Generate Bernoulli trials

plt.hist(data, bins=2, edgecolor='black')  # Histogram with 2 bins for 0 and 1
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution with p=0.6')
plt.axvline(x=bernouilli_mean(p), color='r', linestyle='--', label='Mean')
plt.show()
```

### Poisson distribution
- Parameters:  $\lambda \in (0,\infty)$(Rate), $k\in \{0,1,...\}$.
- PDF : $\displaystyle P(X=k)=\frac{\lambda^k}{k!}e^{-\lambda}⇒P(x)=e^{-\lambda}\sum_{i\geq 0}\frac{\lambda^i}{i!}\chi_i (x)$.
- CDF : $\displaystyle P(X\leq x)=e^{-\lambda}\sum_{k=0}^{\lfloor x ⌋}\frac{\lambda^k}{k!}$.   
-mean
$$\mathbb{E}[X]=\sum_{k=0}^\infty kP(X=k)=\sum_{k=0}^\infty k\frac{\lambda^k}{k!}e^{-\lambda}=e^{-\lambda}\lambda \sum_{k=0}^\infty \frac{\lambda^k}{k!}=\lambda$$
- Variance
$$Var(X)=\mathbb{E}[X^2]-\mathbb{E}[X]^2=\sum_{k=0}^\infty k^2P(X=k) -\lambda^2=\sum_{k=0}^\infty k^2\frac{\lambda^k}{k!}e^{-\lambda}-\lambda^2=e^{-\lambda}\lambda \left(\sum_{k=0}^\infty k\frac{\lambda^k}{k!}+\sum_{k=0}^\infty \frac{\lambda^k}{k!} \right)-\lambda^2=e^{-\lambda}\lambda(e^\lambda\lambda +e^\lambda)-\lambda^2=\lambda$$

- Skewness :
$$\frac{\mu_3}{\sigma^3}=\sum_{k=0}^∞ \biggl(\frac{k-λ}{\sqrt{\lambda}} \biggl)^3\frac{\lambda^k}{k!}e^{-\lambda}=\frac{e^{-\lambda}}{\lambda\sqrt{\lambda}}\biggl(\sum_{k=0}^∞ (k^3-3\lambda k^2+3\lambda^2 k-\lambda^3)\frac{\lambda^k}{k!} \biggr)=\frac{\kappa_3}{\lambda \sqrt{\lambda}}=\frac{1}{\sqrt{\lambda}}$$
- Kurtosis:
$$\frac{\mu_4}{\sigma^4}=\sum_{k=0}^∞ \biggl(\frac{k-λ}{\sqrt{\lambda}} \biggl)^4\frac{\lambda^k}{k!}e^{-\lambda}=\frac{\kappa_4}{\lambda^2}+3\frac{κ_2^2}{\lambda^2}=1+\frac{3}{\lambda}$$

- Characteristic function:

$$φ_X(t)=\int_{\mathbb{R}}e^{itx}P(x)dx=e^{-\lambda}\int_{\mathbb{R}}e^{itx}\sum_{k\geq 0}\frac{\lambda ^k}{k!}\chi_k (x)dx=e^{-\lambda}\sum_{k\geq 0}\frac{\lambda ^k}{k!}\int_{\mathbb{R}}e^{itx}\chi_k (x)dx=e^{-\lambda}\sum_{k\geq 0}\frac{\lambda ^k}{k!}e^{itk}=\exp{(-\lambda+\lambda e^{it})}$$
- cumulant-generative function and cumulants:
$$H(t)=\log{(φ_X(t))}=-\lambda +\lambda e^{it} =\sum_{n\geq 0}κ_n \frac{(it)^n}{n!} ⇒ \kappa_n=\lambda \; \;  \forall n\in \mathbb{N}$$


```python
def poisson_pdf(k,Lambda): #k >=0, lambda>0

  return np.exp(-Lambda)*Lambda**k/math.factorial(k)
def poisson_cdf(x,Lambda): #x : one value , Lambda: one value
  if x<0:
    return 0
  else: #x>=0
    summ=1
    x=int(x)
    for i in range(x,0,-1):
       summ=1+summ*Lambda/i
    summ=np.exp(-Lambda)*summ
    return summ
def poisson_cdf_Array(X,L):
  return np.array([[poisson_cdf(x,l) for x in X] for l in L])

def poisson_mean(Lambda):
  return Lambda
def poisson_variance(Lambda):
  return Lambda
def poisson_skewness(Lambda):
  return 1/np.sqrt(Lambda)
def poisson_kurtosis(Lambda):
  return 1+3/Lambda
Lambda=2
print("poisson distribution for lambda="+str(Lambda))
print("mean is ", poisson_mean(Lambda))
print("variance is ", poisson_variance(Lambda))
print("skewness is ", poisson_skewness(Lambda))
print("kurtosis is ", poisson_kurtosis(Lambda))
#plotting of library and the above function
x=np.linspace(0,20,20,endpoint=False,dtype=np.int32)
y1=np.array([poisson_pdf(xx,Lambda) for xx in x])
y2=np.array([poisson_cdf(xx,Lambda) for xx in x])
plt.fill_between(x,y1,color="red",label="pdf")
plt.plot(x,y2,"o",color="blue",label="cdf")
plt.axvline(x=poisson_mean(Lambda), color='green', linestyle='--', label='Mean')
plt.axvline(x=poisson_mean(Lambda) + np.sqrt(poisson_variance(Lambda)), color='orange', linestyle='--', label='Mean + Std Dev')
plt.axvline(x=poisson_mean(Lambda) - np.sqrt(poisson_variance(Lambda)), color='orange', linestyle='--', label='Mean - Std Dev')
plt.legend()
plt.xlabel("k")
plt.ylabel("probability")
plt.title("Poisson distrubtion, lambda="+str(Lambda))
plt.show()

Lambda=0.5
print("poisson distribution for lambda="+str(Lambda))
print("mean is ", poisson_mean(Lambda))
print("variance is ", poisson_variance(Lambda))
print("skewness is ", poisson_skewness(Lambda))
print("kurtosis is ", poisson_kurtosis(Lambda))
#plotting of library and the above function
x=np.linspace(0,20,20,endpoint=False,dtype=np.int32)
y1=np.array([poisson_pdf(xx,Lambda) for xx in x])
y2=np.array([poisson_cdf(xx,Lambda) for xx in x])
plt.fill_between(x,y1,color="red",label="pdf")
plt.plot(x,y2,"o",color="blue",label="cdf")
plt.axvline(x=poisson_mean(Lambda), color='green', linestyle='--', label='Mean')
plt.axvline(x=poisson_mean(Lambda) + np.sqrt(poisson_variance(Lambda)), color='orange', linestyle='--', label='Mean + Std Dev')
plt.axvline(x=poisson_mean(Lambda) - np.sqrt(poisson_variance(Lambda)), color='orange', linestyle='--', label='Mean - Std Dev')
plt.legend()
plt.xlabel("k")
plt.ylabel("probability")
plt.title("Poisson distrubtion, lambda="+str(Lambda))
plt.show()
```

### Gaussian distribution
- Parameters : $\mu$ (mean) and $\sigma^2 >0$ (variance) .  
- PDF: $\displaystyle P(x|\mu,\sigma^2)=\frac{1}{\sqrt{2\pi\sigma^2}}\exp{\left( -\frac{(x-\mu)^2}{2\sigma^2}\right)}$.  
- CDF: $\displaystyle P(X\leq x|\mu,\sigma)=\frac{1}{\sqrt{2\pi\sigma^2}}\int_{-∞}^x\exp{\left( -\frac{(t-\mu)^2}{2\sigma^2}\right)dt}$. After changing the variable $u=(t-\mu)/\sqrt{2\sigma^2}$, we have : $\displaystyle P(X\leq x|\mu,\sigma)=\frac{1}{\sqrt{\pi}}\int_{-∞}^{(x-\mu)/\sqrt{2\sigma^2}}\exp{\left( -u^2\right)du}=1-\frac{1}{\sqrt{\pi}}\left( \int_{0}^∞-\frac{1}{2}\int_{-(x-\mu)/\sqrt{2\sigma^2}}^{(x-\mu)/\sqrt{2\sigma^2}}\right)[\exp{(-u^2)}du]=\frac{1}{2}(1+\text{Erf}((x-\mu)/\sqrt{2\sigma^2}))$.  
Where the  [error function](https://en.wikipedia.org/wiki/Error_function) is : $\displaystyle \text{Erf}(x)=\int_{-x}^x\exp{-u^2}du ≈ 1-(a_1t+...+a_5t^5)e^{-x^2}$ where $p = 0.3275911, a_1 = 0.254829592, a_2 = −0.284496736, a_3 = 1.421413741, a_4 = −1.453152027, a_5 = 1.061405429, \displaystyle t=\frac{1}{1+px}$.  

- Mean : $\displaystyle \mathbb{E}[X]=\int_{\mathbb{R}}xP(x|\mu,\sigma^2)dx=\frac{1}{\sqrt{\pi}}\left(\int_{\mathbb{R}}u\sqrt{2\sigma^2}\exp{(-u^2)}du+\mu\int_{\mathbb{R}}\exp{(-u^2)}du \right)=\mu$.  

- Variance: $\displaystyle Var(X)=\mathbb{E}[X^2]-\mathbb{E}[X]^2=\int_{\mathbb{R}}x^2P(X\leq x|\mu,\sigma^2)dx-\mu^2=\frac{1}{\sqrt{\pi}}\left(2\sigma^2\int_{\mathbb{R}}u^2\exp{(-u^2)}du+\mu^2\int_{\mathbb{R}}\exp{(-u^2)}du \right)-\mu^2=\frac{2\sigma^2}{\sqrt{\pi}}\left(0+\frac{1}{2}\int_{\mathbb{R}}\exp{(-u^2)}du \right)=\sigma^2$.  
- Skewness:
$$\frac{\mu_3}{σ^3}=\frac{κ_3}{σ^3}=0 $$
- Kurtosis:
$$\frac{\mu_4}{σ^4}=\frac{κ_4}{σ^4}+3\frac{κ_2^2}{σ^4}=3 $$

- Characteristic function:
$$φ_X(t)=\frac{1}{\sqrt{2\pi\sigma^2}}∫_{\mathbb{R}}e^{itx}\exp{\left( -\frac{(x-\mu)^2}{2\sigma^2}\right)}dx=\frac{1}{\sqrt{2\pi\sigma^2}}∫_{\mathbb{R}}\exp{\left( -\frac{(x-\mu-itσ^2)^2}{2\sigma^2}\right) \exp{(-\frac{1}{2}\sigma^2t^2+itμ)}}dx=\exp{(\frac{1}{2}\sigma^2(it)^2+itμ)}=\exp{\left(\frac{1}{2}(σit+\frac{μ}{\sigma})^2\right)}\exp{\left(-\frac{1}{2}\frac{μ^2}{σ^2}\right)}$$

- Cumulants and cumulant-generating function:
$$H(t)=\log{(φ_X(t))}=\frac{1}{2}\sigma^2(it)^2+itμ ⇒ κ_1=μ, κ_2=σ^2, κ_n=0 \; \; ∀n \geq 3
$$


```python
def gaussian_pdf(x,mu,sigma2): #sigma2 is variance
  result=(x-mu)**2/(2*sigma2)
  result=np.exp(-result)
  result=result/np.sqrt(2*np.pi*sigma2)
  return result

#approximation of error function
def erf_appr(x):
  if x.all()==0:
    return 0
  elif x>0:
    p = 0.3275911
    a1 = 0.254829592
    a2 =-0.284496736
    a3 = 1.421413741
    a4 =-1.453152027
    a5 = 1.061405429
    t=1/(1+p*x)
    a=np.array([a1,a2,a3,a4,a5])
    summ=0
    for i in range(4,-1,-1):
      summ=a[i]+t*summ
    summ=summ*t
    summ=summ*np.exp(-x**2)
    return 1-summ
  elif x<0:
    return -erf_appr(-x)

def erf_appr_array(X): #x is an array
  result=np.zeros_like(X)
  for i in range(len(X)):
    x=X[i]
    result[i]=erf_appr(x)
  return result

def gaussian_cdf(x,mu,sigma2):
  x=(x-mu)/np.sqrt(2*sigma2)
  x=erf_appr(x)
  return 0.5*(1+x)
def gaussian_cdf_array(x,mu,sigma2):
  x=(x-mu)/np.sqrt(2*sigma2)
  x=erf_appr_array(x)
  return 0.5*(1+x)

mu,sigma2=0,1
print("gaussian distribution with mean="+str(mu)+" and variance="+str(sigma2))
print("skewness=0 and kurtosis=3")
x=np.linspace(-10,10,1000)
y=gaussian_pdf(x,mu,sigma2)
z=gaussian_cdf_array(x,mu,sigma2)
plt.fill_between(x,y,step="pre",alpha=0.6,label="mean="+str(mu)+" variance ="+str(sigma2))
plt.axvline(x=mu, color='green', linestyle='--', label='Mean')
plt.axvline(x=mu + np.sqrt(sigma2), color='orange', linestyle='--', label='Mean + Std Dev')
plt.axvline(x=mu - np.sqrt(sigma2), color='orange', linestyle='--', label='Mean - Std Dev')
plt.plot(x,z,color="red",label="cdf")
plt.legend()
plt.title("gaussian distribution mean:"+str(mu)+",var:"+str(sigma2))
plt.xlabel("x")
plt.ylabel("p(x)")
plt.show()

mu,sigma2=1,2
print("gaussian distribution with mean="+str(mu)+" and variance="+str(sigma2))
print("skewness=0 and kurtosis=3")
x=np.linspace(-9,11,1000)
y=gaussian_pdf(x,mu,sigma2)
z=gaussian_cdf_array(x,mu,sigma2)
plt.fill_between(x,y,step="pre",alpha=0.6,label="mean="+str(mu)+" variance ="+str(sigma2))
plt.axvline(x=mu, color='green', linestyle='--', label='Mean')
plt.axvline(x=mu + np.sqrt(sigma2), color='orange', linestyle='--', label='Mean + Std Dev')
plt.axvline(x=mu - np.sqrt(sigma2), color='orange', linestyle='--', label='Mean - Std Dev')
plt.plot(x,z,color="red",label="cdf")
plt.legend()
plt.title("gaussian distribution mean:"+str(mu)+",var:"+str(sigma2))
plt.xlabel("x")
plt.ylabel("p(x)")
plt.show()
```

### Exponential distribution
- Parameters : $\lambda>0$, rate.  
PDF: $P(x)=\lambda \exp{(-\lambda x)}, x\geq 0$.  
- CDF: $P(X\leq x)=\displaystyle \int_0^x P(t)dt=1-\exp(-\lambda x)$.  
- Mean: $\mathbb{E}[X]=\displaystyle \int_0^\infty tP(t)dt=\frac{1}{\lambda} \int_0^∞ t^{2-1}\exp(-t)dt=\frac{Γ(2)}{\lambda}=1/\lambda$.   
- Variance: $Var(X)=\mathbb{E}[X^2]-\mathbb{E}[X]^2=\displaystyle \int_0^\infty t^2P(t)dt-\frac{1}{\lambda^2}=\frac{1}{\lambda^2}\left( \int_0^∞ t^{3-1}\exp(-t)dt -1\right)=\frac{Γ(3)-1}{\lambda^2}=1/\lambda^2$.
- Skewness:
$$\frac{\mu_3}{σ^3}=\frac{κ_3}{\sigma^3}=2$$
- Kurtosis:
$$\frac{\mu_4}{σ^4}=\frac{κ_4}{\sigma^4}+3\frac{κ_2^2}{σ^4}=3!+3=9$$

- Characteristic function:
$$φ_X(t)=∫_{0}^\infty e^{itx}λe^{-λx}dx=\frac{λ}{\lambda^2+t^2}\left(λ+it \right)=\frac{λ^2+λ(it)}{\lambda^2-(it)^2}=\frac{1}{1-(it)/λ}$$
- Moment:
$$φ_X(t)=1+\frac{1}{λ}(it)+\frac{2!}{λ^2}\frac{(it)^2}{2!}+... ⇒ ⟨X^n⟩=\frac{n!}{λ^n}$$
- Cumulants and cumulant-generating function:
$$H(t)=∑_{n\geq 0}κ_n \frac{(it)^n}{n!}=\log(φ_X(t))=-\log(1-\frac{it}{λ})=\frac{0!}{λ^1}(it)+\frac{1!}{λ^2}\frac{(it)^2}{2!}+... ⇒ \kappa_n=\frac{(n-1)!}{λ^n}$$




```python
def exponential_pdf(x,Lambda):
  return Lambda*np.exp(-Lambda*x)
def exponential_cdf(x,Lambda):
  return 1-np.exp(-Lambda*x)
def exponential_mean(Lambda):
  return 1/Lambda
def exponential_variance(Lambda):
  return 1/Lambda**2
Lambda=0.5
print("exponential distribution with λ=",Lambda)
print("mean=",exponential_mean(Lambda))
print("variance=",exponential_variance(Lambda))
print("skewness=",2)
print("kurtosis=",9)

x=np.linspace(0,10,1000)
y=exponential_pdf(x,Lambda)
z=exponential_cdf(x,Lambda)
plt.fill_between(x,y,step="pre",alpha=0.5,label="pdf")
plt.plot(x,z,label="cdf")
plt.axvline(x=exponential_mean(Lambda), color='green', linestyle='--', label='Mean')
plt.axvline(x=exponential_mean(Lambda) + np.sqrt(exponential_variance(Lambda)), color='orange', linestyle='--', label='Mean + Std Dev')
plt.axvline(x=exponential_mean(Lambda) - np.sqrt(exponential_variance(Lambda)), color='orange', linestyle='--', label='Mean - Std Dev')
plt.legend()
plt.title("exponential distribution with λ="+str(Lambda))
plt.xlabel("x")
plt.ylabel("P(x)")
plt.show()

Lambda=2
print("exponential distribution with λ="+str(Lambda))
print("mean=",exponential_mean(Lambda))
print("variance=",exponential_variance(Lambda))
print("skewness=",2)
print("kurtosis=",9)

x=np.linspace(0,5,1000)
y=exponential_pdf(x,Lambda)
z=exponential_cdf(x,Lambda)
plt.fill_between(x,y,step="pre",alpha=0.5,label="pdf")
plt.plot(x,z,label="cdf")
plt.axvline(x=exponential_mean(Lambda), color='green', linestyle='--', label='Mean')
plt.axvline(x=exponential_mean(Lambda) + np.sqrt(exponential_variance(Lambda)), color='orange', linestyle='--', label='Mean + Std Dev')
plt.axvline(x=exponential_mean(Lambda) - np.sqrt(exponential_variance(Lambda)), color='orange', linestyle='--', label='Mean - Std Dev')
plt.legend()
plt.title("exponential distribution with λ="+str(Lambda))
plt.xlabel("x")
plt.ylabel("P(x)")
plt.show()
```

### Cauchy-Lorentz distribution
- Parameters : $x_0$ (location) and $\gamma>0$ (scale)
- PDF : $\displaystyle P(x)=\frac{1}{\pi γ}\frac{1}{1+\left(\frac{x-x_0}{γ} \right)^2}, x\in \mathbb{R}$.
- CDF : $\displaystyle P(X\leq x)=\frac{1}{\pi γ}\int_{-\infty}^x\frac{dt}{1+\left(\frac{t-x_0}{γ} \right)^2}=\frac{1}{\pi }\int_{-\infty}^{\frac{x-x_0}{γ} }\frac{du}{1+u^2}=\frac{1}{π}\arctan{(\frac{x-x_0}{γ})}+\frac{1}{2}$.
- Mean : $\displaystyle \mathbb{E}[X]=\int_{-∞}^\infty xP(x)dx=\frac{1}{\pi }\int_{-\infty}^∞\frac{x_0+\gamma u}{1+u^2}du$ which diverges. So mean is undefined.  
- Variance : similarly as in mean, variance is also
 undefined.

- Characteristic function:

 $$φ_X(t)=\frac{1}{\pi γ}∫_{\mathbb{R}}e^{ixt}\frac{dx}{1+\left(\frac{x-x_0}{γ} \right)^2}=\frac{1}{\pi }e^{ix_0t}∫_{\mathbb{R}}\frac{\exp{(iγut)}}{1+u^2}du=\frac{1}{\pi }e^{ix_0t}∫_{\mathbb{R}}\frac{\exp{(iγut)}}{1+u^2}du=\frac{1}{\pi }e^{ix_0t}\left(2πi\exp{(-|γt|)}/2i -0 \right)=\exp{(ix_0t-|γt|)}$$

 - cumulants and cumulant-generating function:
 $\displaystyle H(t)=\log{(φ_X(t))}=x_0 (it)-|γ||t|⇒ κ_n $ is not defined as $|t|$ is not smooth in 0.


```python
def cauchy_pdf(x,x0,gamma):
  x=(x-x0)/gamma
  x=1+x**2
  x=1/x
  return (1/np.pi )*(1/gamma)*x
def cauchy_cdf(x,x0,gamma):
  x=(x-x0)/gamma
  x=np.arctan(x)
  return x/np.pi+0.5
X0=[-1,0,1]
Gamma=[0.5,1,2]
x=np.linspace(-10,10,1000)
for x0 in X0:
  for gamma in Gamma:
    y=cauchy_pdf(x,x0,gamma)
    plt.plot(x,y,label="x0="+str(x0)+",γ="+str(gamma))
plt.legend()
plt.title("pdf of Cauchy-Lorentz distribution")
plt.xlabel("x")
plt.ylabel("P(x)")
plt.show()

for x0 in X0:
  for gamma in Gamma:
    y=cauchy_cdf(x,x0,gamma)
    plt.plot(x,y,label="x0="+str(x0)+",γ="+str(gamma))
plt.legend()
plt.title("cdf of Cauchy-Lorentz distribution")
plt.xlabel("x")
plt.ylabel("P(X≤x)")
plt.show()
```

### Continuous Uniform Distribution
- Parameters : $a < b \in \mathbb{R}$.
- PDF : $\displaystyle P(x)=\frac{1}{b-a}\chi_{[a,b]}(x)$,
- CDF : $\displaystyle P(X\leq x)=\frac{1}{b-a}\int_{-\infty}^x χ_{[a,b]}(t)dt=\frac{x-a}{b-a}χ_{[a,b]}(x)+χ_{(b,\infty)}(x)$ .
- Mean : $\displaystyle \mathbb{E}[X]=\int_{-∞}^\infty xP(x)dx=\frac{1}{b-a }\int_{-\infty}^∞{xχ_{[a,b]}(x)dx}= \frac{1}{b-a }\int_{-}^b{xdx}=\frac{a+b}{2}$.
- Variance : $\displaystyle Var(X)=\mathbb{E}[X^2]-\mathbb{E}[X]^2=\int_{-∞}^\infty x^2P(x)dx-\frac{a^2+b^2+2ab}{4}=\frac{1}{b-a }\int_{a}^b{x^2dx}-\frac{a^2+b^2+2ab}{4}= \frac{a^2+b^2+ab}{3}-\frac{a^2+b^2+2ab}{4}=\frac{(a-b)^2}{12} $.
- Skewness:
$$\frac{\mu_3}{\sigma^3}=\frac{1}{\sigma^3(b-a)}\int_{a}^b{(x-\mu)^3dx}=\frac{1}{\sigma^3(b-a)}\int_{-(b-a)/2}^{(b-a)/2}{x^3dx}=0$$
- Kurtosis:
$$\frac{\mu_4}{\sigma^4}=\frac{1}{\sigma^4(b-a)}\int_{a}^b{(x-\mu)^4dx}=\frac{1}{\sigma^4(b-a)}\int_{-(b-a)/2}^{(b-a)/2}{x^4dx}=\frac{2}{5}\frac{12^2}{2^5}=9/5$$
- Characteristic function:
$$φ_X(t)=\frac{1}{b-a}∫_{\mathbb{R}}e^{itx}\chi_{[a,b]}(x)dx=\frac{1}{b-a}\frac{\exp(itb)-\exp(ita)}{it}$$


```python
def uniform_pdf(x,a,b): #a<b
  if x<a or x>b:
    return 0
  else:
    return 1/(b-a)

def uniform_cdf(x,a,b):
  if x<=a :
    return 0
  else:
    if x>=b:
      return 1
    else:
      return (x-a)/(b-a)

def uniform_array(x,a,b,arg):
  result=np.zeros_like(x)
  if arg=="pdf":
    for i in range(len(x)):
      result[i]=uniform_pdf(x[i],a,b)
  elif arg=="cdf":
    for i in range(len(x)):
      result[i]=uniform_cdf(x[i],a,b)
  return result
def uniform_mean(a,b):
  return (a+b)/2
def uniform_variance(a,b):
  return (a-b)**2/12
a,b=-2,4
x=np.linspace(-3,5,10000)
y=uniform_array(x,a,b,"pdf")
z=uniform_array(x,a,b,"cdf")

plt.fill_between(x,y,step="pre",alpha=0.6,label=("("+str(a)+","+str(b)+")"))
plt.axvline(x=uniform_mean(a,b), color='green', linestyle='--', label='Mean')
plt.axvline(x=uniform_mean(a,b) + np.sqrt(uniform_variance(a,b)), color='orange', linestyle='--', label='Mean + Std Dev')
plt.axvline(x=uniform_mean(a,b) - np.sqrt(uniform_variance(a,b)), color='orange', linestyle='--', label='Mean - Std Dev')
plt.plot(x,z,color="red",label="cdf")

plt.title("uniform distribution ("+str(a)+","+str(b)+")")
plt.legend()
plt.xlabel("x")
plt.ylabel("P(x)")
plt.show()
a,b=0,1
x=np.linspace(-1,2,1000)
y=uniform_array(x,a,b,"pdf")
z=uniform_array(x,a,b,"cdf")

plt.fill_between(x,y,step="pre",alpha=0.6,label=("("+str(a)+","+str(b)+")"))
plt.axvline(x=uniform_mean(a,b), color='green', linestyle='--', label='Mean')
plt.axvline(x=uniform_mean(a,b) + np.sqrt(uniform_variance(a,b)), color='orange', linestyle='--', label='Mean + Std Dev')
plt.axvline(x=uniform_mean(a,b) - np.sqrt(uniform_variance(a,b)), color='orange', linestyle='--', label='Mean - Std Dev')
plt.plot(x,z,color="red",label="cdf")

plt.title("uniform distribution ("+str(a)+","+str(b)+")")
plt.legend()
plt.xlabel("x")
plt.ylabel("P(x)")
plt.show()
```

## Probabilistic inequalities
### Markov's Inequality
$$P(X\geq a)\leq \frac{\mathbb{E}[X]}{a} \; \; \forall a> 0,X\geq 0$$
**Proof**:  
Let $X\geq 0,a>0$. Then we have
$$\mathbb{E}[X]=\int_{\mathbb{R}}xP(x)dx=\int_{a}^∞xP(x)dx+\int_{0}^axP(x)dx \geq \int_{a}^∞xP(x)\geq \int_{a}^∞aP(x)=aP(X\geq a)$$
### Chebyshev's Inequality
$$P(|X-\mathbb{E}[X]|\geq σa)\leq\frac{1}{a^2} \;\;\forall a,σ^2>0, $$
**Proof**: Let a,σ^2>0, From Markov's Inequality:  

$$P(|X-\mathbb{E}[X]|\geq σa)=P((X-\mathbb{E}[X])^2\geq σ^2a^2)\leq \frac{\mathbb{E}[(X-\mathbb{E}[X])^2]}{σ^2a^2}=\frac{1}{a^2}$$

### Chernoff bound
$$P(x\geq a)=P(e^{tx}\geq e^{at})\leq \frac{\mathbb{E}[e^{tx}]}{e^{ta}}$$

### Example-Problem : Coupon collector’s problem

**Problem** : Assume that there is $n$ different coupons and you want to collect all of them. At every
step you can get only one random coupon. What is the probability that you still do not
have all coupons after $t$ steps?  
**Answer** :

For 1 step: The probability of not getting one particular coupon is $\frac{n-1}{n}$. Hence for $t$ steps, this probability of not getting one particular coupon is $(\frac{n-1}{n})^t$. So there are on average $n(\frac{n-1}{n})^t$ coupon missing after $t$ steps. If $Y=$ number of missing coupon, then :   
$\displaystyle P(Y\geq 1)\leq \frac{n(1-1/n)^t}{1} \leq n\exp{(-t/n)}$.   

To find the average number of purchase to get all coupon, then consider the random variables $X_1,...,X_n$ such that :    
$X_1: $ purchase needed for getting 1 coupon, $X_i (i\geq 2):$ purchase needed of getting one distinct coupon after having $i-1$ distinct coupon.  Then $X=X_1+...+X_n$ is the purchase needed for getting all the coupon. $X_1,...,X_n$ follow geometric distribution (getting the 1st success after loss of the previous) with probability $p_i$ of $X_i$ is $p_i=\frac{n-(i-1)}{n}$ since we already have $i-1$ coupon and it is the probability of taking one from the other $n-(i-1)$ coupons with expected value $\mathbb{E}[X_i]=1/p_i$.
$$\mathbb{E}[X]=\mathbb{E}[X_1+...+X_n]=\mathbb{E}[X_1]+...+\mathbb{E}[X_n]=\sum_{i=1}^n\frac{n}{n-i+1}=n\left(1+\frac{1}{2}+...+\frac{1}{n} \right)=nH_n≈n\left(\log(n)+γ+\frac{1}{2n}-\frac{1}{12n^2}+\frac{1}{120n^4}+O\left(\frac{1}{n^6}\right)\right)=n\log(n)+γn+\frac{1}{2}-\frac{1}{12n}+\frac{1}{120n^3}+O\left(\frac{1}{n^5}\right)$$




```python
t_values=np.arange(1,100)
n_values=np.arange(1,100)
def f(N, T):
    result = np.ones(N.shape)  # Create an array of ones with the same shape as N
    mask = N * np.exp(-T / N) <= 1  # Create a mask for the condition
    result[mask] = N[mask] * np.exp(-T[mask] / N[mask])  # Apply the condition
    return result

N,T=np.meshgrid(n_values,t_values)
prob=f(N,T)
plt.contourf(N,T,prob,levels=112,cmap="viridis")
plt.colorbar(label="upper bound probability")
plt.title("probability of missing at least one coupon and its upper bound")
plt.xlabel("number of coupon n")
plt.ylabel("number of steps t")
plt.xlim(1,100)
plt.ylim(1,100)
plt.grid()
plt.legend()
plt.show()

def expected_value(n):
    gamma = 0.57721566490153286060651209008240243104215933593992
    return gamma*n + n*np.log(n) + 0.5 - 1./(12*n) + 1./(120*n**3)

def collect_coupons(n):
    coupons = np.arange(1, n+1)
    collected = np.zeros(n, dtype=bool)
    purchases = 0
    while not np.all(collected):
        new_coupon = np.random.choice(coupons)
        collected[new_coupon-1] = True
        purchases += 1
    return purchases

purchases = []
E_X = []
for n in range(1, 100):
    purchase = collect_coupons(n)
    purchases.append(purchase)
    E_X.append(expected_value(n))

plt.plot(purchases, label="experimental purchase needed")
plt.plot(E_X, label="average purchase")

plt.title("purchase needed to obtain all n distincts coupons")
plt.xlabel("n")
plt.ylabel("number of purchase")
plt.show()
```
